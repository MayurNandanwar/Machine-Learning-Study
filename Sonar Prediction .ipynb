```python
# split dataset in train and test
from sklearn.model_selection import train_test_split
```

```python
# seperate x and y
X = df.drop(60,axis=1)
y = df[60]
```

```python
# just assign category available in y to number 
y = y.map({'R':0,'M':1})
```

```python
# split data in to train and test ,only 10% data is going to be in testset
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=10)
```

```python
# this is binary classification problem 
from sklearn.linear_model import LogisticRegression
```

```python
# object of Logistic Regressir
logir = LogisticRegression()
```

```python
# to train model on train dataset to learn params or patters hidden in dataset 
logir.fit(x_train,y_train)
```

```python
y_train_pred = logir.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_recall_fscore_support
```

```python
train_accuracy = accuracy_score(y_train,y_train_pred)
```

```python
from yellowbrick.classifier import precision_recall_curve
```

```python
from yellowbrick.classifier import (
    ClassificationReport,
    DiscriminationThreshold,
    ROCAUC
)
```

```python
visualizer = ClassificationReport(logir)

visualizer.fit(x_train, y_train)        # Fit the visualizer and the model
visualizer.score(x_test, y_test)        # Evaluate the model on the test data
visualizer.show()  

# from this we can see that recall for class 1 is 1 means no mistaken is made by model when prediction observation belongs to 
# class 1 but for class 0 recall is 0.6 means 40 % chances that ti make mistake to predict class 0

# Precision for class 1 is 0.5 means that at time of prediction class 0's observation model makes so many mistack to predict 0 and instead of 
# it predicts 1# means out of 100 obs of class 0 for 50 observation model will predict 1 and
# for observation belongs to class 1 is indeed model predictis 1 model makes no mistake to predict class 1 because recall is 1

# here f1 score is metrix that optimizing value of precision over recall, harmonic mean of recall and precision range between 0 and 1
# near to 0 means worst and near to 1 best
```

```python
visualizer = DiscriminationThreshold(logir,
                                     n_trials=10,
                                     cv=0.1,
                                     argmax='fscore',
                                     random_state=0,
                                     is_fitted='auto',
                                     exclude = "queue_rate")

visualizer.fit(x_train, y_train)        # Fit the visualizer and the model
visualizer.score(x_test, y_test)        # Evaluate the model on the test data
visualizer.show()
```

```python

vi= ROCAUC(estimator=logir,classes=[0,1],)
vi.fit(x_train,y_train)
vi.score(x_test,y_test)
vi.show()
```
