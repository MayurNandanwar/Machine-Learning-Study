```python
# groupBY :- use of pivot
import pandas as pd
```

```python
df = pd.read_csv('for_box_plot.csv')
```

```python
df
```

```python
# box plot:- hourly what is box plot like percentile , 
import seaborn as sns 
sns.set(rc={'figure.figsize':(20,10)})
sns.boxplot(x=df['hh'],y=df['estimated_dt'])
```

```python
A = df.loc[0:5,['trip_master_id','mindiff','estimated_dt']]
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(A,annot=True,cmap='RdBu')
```

```python
grp_df = df[['hh','mi','estimated_dt']].groupby(['hh','mi'],as_index=False).mean()
df_pivot = grp_df.pivot(index='hh',columns='mi') # this shows mean value in index is hh and in column is mi
df_pivot
```

```python
# heat map : used to plot target value over multiple variable check corelation between variable 
import seaborn as sns

sns.set(rc={'figure.figsize':(30,30)})
sns.heatmap(df_pivot,annot=True)
```
